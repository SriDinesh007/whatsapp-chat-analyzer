```python
def month_activity_map(selected_user,df):
    if selected_user != 'Overall':
        df = df[df['users']== selected_user]
    return df['month'].value_counts()
```

```python
def monthtimeline(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['users'] == selected_user]

    temp = df.groupby(['year', 'month_num', 'month']).count()[
        'messages'].reset_index()

    time = []
    for i in range(temp.shape[0]):
        time.append(temp['month'][i]+"-"+str(temp['year'][i]))

    temp['time'] = time

    return temp
```

```python
monthtimeline(selected_user,df)
```

```python
import emoji
def getemojistats(selected_user, df):

    if selected_user != 'Overall':
        df = df[df['users'] == selected_user]

    emojis = []
    for message in df['messages']:
        emojis.extend([c for c in message if emoji.is_emoji(c)])

    emojidf = pd.DataFrame(Counter(emojis).most_common(len(Counter(emojis))))

    return emojidf
```

```python
getemojistats(selected_user,df)
```

```python
# get common words
# This will return a dataframe of most common words
def get_common_words(selected_user,df):
    # getting the stopwords
    with open('stop_hinglish.txt','r',encoding = 'utf-8') as file:
        stopwords = file.read()

    stopwords = stopwords.split('\n')

    if selected_user != 'Overall':
        df = df[df['users']==selected_user]

    temp = df[(df['users'] != 'Group Notification') ]

    words = []
    for message in temp['messages']:
        if message.strip() != 'This message was deleted':
            for word in message.lower().split():
                if word not in stopwords and word not in ['<media','<Media','omitted>']:
                    words.append(word)

    most_common_words_df = pd.DataFrame(Counter(words).most_common())
    return most_common_words_df
    


        
```

```python
get_common_words(selected_user,df)
```

```python
from wordcloud import WordCloud


def create_word_cloud(selected_user, df):

    df = get_common_words(selected_user,df)
    # if selected_user != 'Overall':
    #     df = df[df['users'] == selected_user]

    wc = WordCloud(width=500, height=500,
                   min_font_size=10, background_color='white')

    df_wc = wc.generate(df[0].str.cat(sep=" "))

    return df_wc
```

```python
def fetch_busy_user(df):

    df = df[df['users'] != 'Group Notification']
    count = df['users'].value_counts().head()

    newdf = pd.DataFrame((df['users'].value_counts()/df.shape[0])*100)
    newdf.columns = ['percent']  # Optional but improves clarity
    newdf = newdf.reset_index().rename(columns={'index': 'User'})

    return count, newdf
```

```python
from urlextract import URLExtract
extract = URLExtract()
def fetch_stats(selected_user,df):
    if selected_user != 'Overall':
        df = df[df['users']==selected_user]

    num_messages = df.shape[0]
    words = []
    for message in df['messages']:
        words.extend(message.split())

    # counting the number of media files shared
    media_omitted = df[df['messages'].str.strip().isin(['<Media omitted>','<media omitted>'])] 

    # no of links shared
    links = []
    for message in df['messages']:
        links.extend(extract.find_urls(message))

    return num_messages, len(words), media_omitted.shape[0], len(links)
```

```python
x, y, z, w = fetch_stats(selected_user,df)
```
